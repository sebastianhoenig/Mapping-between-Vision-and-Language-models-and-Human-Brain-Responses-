```python
def plot_spearmanr_reverse(df_sub, title, ax, ncol, bbox):
    x_values = ['EVC', 'LOC', 'PFS', 'OPA', 'PPA', 'RSC']

    plt.style.use("seaborn-v0_8-whitegrid")

    handles = []
    for i, column in enumerate(df_sub.columns):
        line, = ax.plot(x_values, df_sub[column], marker='o', label=column)
        handles.append(line)

    # Create the color-coded legend
    ax.legend(handles=handles, labels=df_sub.columns.tolist(), loc='upper right', ncol=ncol, fontsize='xx-small', bbox_to_anchor=(bbox, 1))
    ax.set_title(f'{title} Layers', fontsize=20, fontweight='bold')
```

```python
fig, axs = plt.subplots(nrows=6, ncols=1, figsize=(7, 21), sharey=True, gridspec_kw={'wspace': 0.5, 'hspace': 0.5})
plot_spearmanr_reverse(df.iloc[:6, 6:18], "CLIP Image", axs[0], 2, 1.43)
plot_spearmanr_reverse(df.iloc[:6, 18:30], "CLIP Text", axs[1], 2, 1.4)
plot_spearmanr_reverse(df.iloc[:6, 30:42], "Vilt Multi", axs[2], 2, 1.32)
plot_spearmanr_reverse(df.iloc[:6, 42:54], "ALBEF Image", axs[3], 2, 1.46)
plot_spearmanr_reverse(df.iloc[:6, 54:60], "ALBEF Multi", axs[4], 1, 1.21)
plot_spearmanr_reverse(df.iloc[:6, 60:66], "ALBEF Text", axs[5], 1, 1.21)
plt.show()
```
